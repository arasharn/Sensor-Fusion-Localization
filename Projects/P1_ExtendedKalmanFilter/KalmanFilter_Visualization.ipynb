```python
import pandas as pd
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
def loading_data(file_path):
    df = pd.read_csv(file_path, sep="\t", header=None)
    est_px = []
    est_py = []
    est_vx = []
    est_vy = []

    meas_px = []
    meas_py = []

    gt_px = []
    gt_py = []
    gt_vx = []
    gt_vy = []
    
    [est_px.append(df.get_value(i, 0)) for i in range(0, len(df))]
    [est_py.append(df.get_value(i, 1)) for i in range(0, len(df))]
    [est_vx.append(df.get_value(i, 2)) for i in range(0, len(df))]
    [est_vy.append(df.get_value(i, 3)) for i in range(0, len(df))]
    
    [meas_px.append(df.get_value(i, 4)) for i in range(0, len(df))]
    [meas_py.append(df.get_value(i, 5)) for i in range(0, len(df))]
    
    [gt_px.append(df.get_value(i, 6)) for i in range(0, len(df))]
    [gt_py.append(df.get_value(i, 7)) for i in range(0, len(df))]
    [gt_vx.append(df.get_value(i, 8)) for i in range(0, len(df))]
    [gt_vy.append(df.get_value(i, 9)) for i in range(0, len(df))]
    
    return (est_px, est_py, est_vx, est_vy, meas_px, meas_py, gt_px, gt_py, gt_vx, gt_vy)
    
```

```python
def plot_position_results(file_path, data_num):
    (est_px, est_py, est_vx, est_vy, meas_px, meas_py, gt_px, gt_py, gt_vx, gt_vy) = loading_data(file_path)
    fig, ax = plt.subplots(figsize=(20,15))
    ax.scatter(est_px, est_py, c='green', label='Kalman Filter', s = 100, alpha=1, edgecolors='none')
    ax.scatter(meas_px, meas_py, c='red', label='Measurements', s = 50, alpha=0.8, edgecolors='none')
    ax.scatter(gt_px, gt_py, c='blue', label='Ground Truth', s = 50, alpha=0.5, edgecolors='none')
    plt.xlabel('x', fontsize=30)
    plt.ylabel('y', fontsize=30)
    ax.legend(fontsize=24,loc='upper left')
    plt.tick_params(axis='both', which='major', labelsize=20)
    ax.grid(True)
    plt.suptitle("Positions for Dataset {}".format(data_num), fontsize=30)
    plt.savefig('./Figures/Dataset{}.jpg'.format(data_num))
    plt.show()
```

```python
filename1 = "obj_pose-laser-radar-synthetic-output.txt"
path1 = "./data/" + filename1

filename2 = "sample-laser-radar-measurement-data-1-output.txt"
path2 = "./data/" + filename2

filename3 = "sample-laser-radar-measurement-data-2-output.txt"
path3 = "./data/" + filename3
```

```python
plot_position_results(path1, filename1)
```

```python
plot_position_results(path2, filename2)
```

```python
plot_position_results(path3, filename3)
```

```python
def plot_velocity_results(file_path, data_num):
    (est_px, est_py, est_vx, est_vy, meas_px, meas_py, gt_px, gt_py, gt_vx, gt_vy) = loading_data(file_path)
    fig, ax = plt.subplots(figsize=(20,15))
    ax.scatter(est_vx, est_vy, c='green', label='Kalman Filter', s = 100, alpha=1, edgecolors='none')
    ax.scatter(gt_vx, gt_vy, c='blue', label='Ground Truth', s = 50, alpha=0.5, edgecolors='none')
    plt.xlabel('vx', fontsize=30)
    plt.ylabel('vy', fontsize=30)
    ax.legend(fontsize=24,loc='upper left')
    plt.tick_params(axis='both', which='major', labelsize=20)
    ax.grid(True)
    plt.suptitle("Velocities for Dataset {}".format(data_num), fontsize=30)
    plt.savefig('./Figures/Velocity_Dataset{}.jpg'.format(data_num))
    plt.show()
```

```python
plot_velocity_results(path1,filename1)
```

```python
plot_velocity_results(path2,filename2)
```

```python
plot_velocity_results(path3,filename3)
```
