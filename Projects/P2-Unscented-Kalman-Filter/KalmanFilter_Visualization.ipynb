```python
import pandas as pd
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
# time_stamp	px_state	py_state	v_state	yaw_angle_state	yaw_rate_state	sensor_type	NIS	px_measured	
#  py_measured	px_ground_truth	py_ground_truth	vx_ground_truth	vy_ground_truth

def loading_data(file_path):
    df = pd.read_csv(file_path, sep="\t", header=None)
    
    time_stamp = []
    est_px = []
    est_py = []
    
    v_state = []
    
    yaw_angle_state = []
    yaw_rate_state = []
    
    sensor = []
    NIS = []
    
    meas_px = []
    meas_py = []

    gt_px = []
    gt_py = []
    gt_vx = []
    gt_vy = []
    
    [time_stamp.append(df.get_value(i, 0)) for i in range(1, len(df))]
    
    [est_px.append(df.get_value(i, 1)) for i in range(1, len(df))]
    [est_py.append(df.get_value(i, 2)) for i in range(1, len(df))]
    
    [v_state.append(df.get_value(i, 3)) for i in range(1, len(df))]
    
    [yaw_angle_state.append(df.get_value(i, 4)) for i in range(1, len(df))]
    [yaw_rate_state.append(df.get_value(i, 5)) for i in range(1, len(df))]

    [sensor.append(df.get_value(i, 6)) for i in range(1, len(df))]
    [NIS.append(df.get_value(i, 7)) for i in range(1, len(df))]
    
    [meas_px.append(df.get_value(i, 8)) for i in range(1, len(df))]
    [meas_py.append(df.get_value(i, 9)) for i in range(1, len(df))]
    
    [gt_px.append(df.get_value(i, 10)) for i in range(1, len(df))]
    [gt_py.append(df.get_value(i, 11)) for i in range(1, len(df))]
    [gt_vx.append(df.get_value(i, 12)) for i in range(1, len(df))]
    [gt_vy.append(df.get_value(i, 13)) for i in range(1, len(df))]
    
    return (time_stamp, est_px, est_py, v_state, yaw_angle_state, yaw_rate_state,
            sensor, NIS, meas_px, meas_py, gt_px, gt_py, gt_vx, gt_vy)
    
```

```python
def plot_position_results(file_path, data_num):
    (time_stamp, est_px, est_py, v_state, yaw_angle_state, yaw_rate_state,
            sensor, NIS, meas_px, meas_py, gt_px, gt_py, gt_vx, gt_vy) = loading_data(file_path)
    fig, ax = plt.subplots(figsize=(20,15))
    ax.scatter(est_px, est_py, c='green', label='Unscented Kalman Filter', s = 100, alpha=1, edgecolors='none')
    ax.scatter(meas_px, meas_py, c='red', label='Measurements', s = 50, alpha=0.8, edgecolors='none')
    ax.scatter(gt_px, gt_py, c='blue', label='Ground Truth', s = 50, alpha=0.5, edgecolors='none')
    plt.xlabel('x', fontsize=30)
    plt.ylabel('y', fontsize=30)
    ax.legend(fontsize=24,loc='upper left')
    plt.tick_params(axis='both', which='major', labelsize=20)
    ax.grid(True)
    plt.suptitle("Positions for Dataset {}".format(data_num), fontsize=30)
    plt.savefig('./Figures/Dataset{}.jpg'.format(data_num))
    plt.show()
```

```python
filename1 = "obj_pose-laser-radar-synthetic-output.txt"
path1 = "./data/" + filename1
```

```python
plot_position_results(path1, filename1)
```
